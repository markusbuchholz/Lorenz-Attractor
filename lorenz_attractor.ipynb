
# Lorenz Attractor


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dt = 0.01
num_steps = 3000
s = 10
r = 28
b = 2.667

def computeLorenz(x0, y0, z0):
    
    
    xX = np.empty(num_steps)
    yY = np.empty(num_steps)
    zZ = np.empty(num_steps)
    
    xX[0] = x0
    yY[0] = y0
    zZ[0] = z0
    
    for i in range (1,num_steps):
        
        xX[i] = xX[i-1] + s*(yY[i-1] - xX[i-1])*dt
        yY[i] = yY[i-1] + (r*xX[i-1] - yY[i-1] - xX[i-1]*zZ[i-1])*dt
        zZ[i] = zZ[i-1] + (xX[i-1]*yY[i-1] - b*zZ[i-1])*dt
                 
    return xX, yY, zZ


x1, y1, z1 = computeLorenz(0,1,2.001)
x2, y2, z2 = computeLorenz(0,1,1)
```

```python
plt.figure(figsize=(16, 16)).add_subplot(projection='3d')

plt.plot(x1, y1, z1)
plt.plot(x2, y2, z2)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.title("Lorenz Attractor")

plt.show()
```

```python
x1, y1, z1 = computeLorenz(0,1,1.05)
x2, y2, z2 = computeLorenz(0,1,1)

t = []
diff_x = []
diff_y = []
diff_z = []
for i in range(len(x1)):
    t.append(i)
    diff_x.append(x1[i] - x2[i])
    diff_y.append(y1[i] - y2[i])
    diff_z.append(z1[i] - z2[i])
    
```

```python
plt.figure(figsize=(12, 4))
plt.plot(t, diff_x)
plt.xlabel("time")
plt.ylabel("difference")
plt.title("Lorenz Attractor: X difference")
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(t, diff_y)
plt.xlabel("time")
plt.ylabel("difference")
plt.title("Lorenz Attractor: Y difference")
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(t, diff_z)
plt.xlabel("time")
plt.ylabel("difference")
plt.title("Lorenz Attractor: Y difference")
plt.show()
```
